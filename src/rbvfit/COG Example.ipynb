# Example Notebook Showing how to create and plot the curve of growth

```python
from rbvfit import compute_cog as c
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
#create a series of column densities for which COG is computed
Nlist=np.arange(11.5,20.,.1)

#create b value lists for which COG is computed
blist=np.array([15.,20.,30.,50.])

#Wavelength of transition in angstrom
lam_guess= 2600#2796.3

#create the COG object
s=c.compute_cog(lam_guess,Nlist,blist)
```

```python
#plot the computed COG

s.plot_cog()
```

```python
# Extract values from object

# Matrix of EWs 
Wlist=s.Wlist


#recreate the plot above

#Convert Angstrom to cm
plt.title(s.st['name'])
for i in range (0,len(blist)):
    plt.plot(np.log10((10**Nlist)*s.st['fval']*s.st['wave']*1e-8),np.log10(Wlist[:,i]/s.st['wave']),label='b = '+ np.str(blist[i]))
plt.xlabel(r'$log_{10} [N f \lambda]$')
plt.ylabel(r'$log_{10} [W/ \lambda]$')

plt.legend()
```
